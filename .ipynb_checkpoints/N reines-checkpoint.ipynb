```python
import numpy as np
from tkinter import *
import tkinter.ttk as ttk
global N 
S=input()
N=int(S)
global M
def affichage(matrice): 
    for i in range(N): 
        print ("____________________") 
        for j in range(N): 
            print (matrice[i][j], "|",end = " ") 
        print()  

def diag(echec,x,y):
    arr=[]
    
    for i in range(N) :
        
        if i != x : 
            if((y-abs(x-i))>=0) :
                
                arr.append(echec[i][y-abs(x-i)])
                
            if((y+abs(x-i))<N):
                
                arr.append(echec[i][y+abs(x-i)])
    return arr 

def placer_reine(matrice, l, c): 
  
    
    for i in range(c): 
        if matrice[l][i] == 1: 
            return False
        
      
    arr=diag(matrice, l, c) 
    n=len(arr)
    
    for i in range(n) : 
        if arr[i]==1 :
            return False
    matrice[l][c] = 1
    return True

def placer_toutes_reines(matrice, c): 
      
 
    if c >= N: 
        return True

    for i in range(N): 
  
        if placer_reine(matrice, i, c): 
              
            if placer_toutes_reines(matrice, c + 1) == True: 
                return True
 
            matrice[i][c] = 0

    return False
   
def programe_principale(): 
    
    matrice = np.zeros( (N, N) , dtype=int)
    
    if placer_toutes_reines(matrice, 0) == False: 
        print ("Pas des solution") 
        return False
    
    affichage(matrice) 
    return True

programe_principale() 

```

```python
matrice = np.zeros( (N, N) , dtype=int)
placer_toutes_reines(matrice,0)
root = Tk()

text = [[None]*N]*N
buttons = [[None]*N]*N

#Create style object
sto = ttk.Style()

#configure style
sto.configure('TButton', font=('calibri', 10, 'bold', 'underline'),foreground='Green')

for i in range(N):
    for j in range(N):
        text[i][j] = StringVar()
        text[i][j].set('(%d)' % matrice[i][j])
        buttons[i][j] = Button(root, command = None)
        if matrice[i][j] == 1 :
            buttons[i][j].config(textvariable = text[i][j], width = 9, height = 5 ,background='light blue', border=10)
        else:
            buttons[i][j].config(textvariable = text[i][j], width = 9, height = 5) 
        buttons[i][j].grid(row = i, column = j)

root.mainloop() 
```
